```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

data = pd.read_csv('clusters.csv')

dates = data['Date'].values
num_patients = data['Random and Returned from other countries'].values

dates = dates[:100]
num_patients = num_patients[:100]

model = LinearRegression()

X = np.arange(len(dates)).reshape(-1, 1)
y = num_patients

model.fit(X, y)

new_date = np.array([len(dates)]).reshape(-1, 1)
predicted_num_patients = model.predict(new_date)

print("Predicted number of patients on 2021-05-20:", predicted_num_patients)
```

```python
import matplotlib.pyplot as plt

plt.scatter(dates, num_patients, color='blue', label='Original Data')
plt.plot(dates, model.predict(X), color='red', label='Linear Regression')
plt.scatter(['2021-05-20'], predicted_num_patients, color='green', label='Predicted Point')
plt.xticks(dates[::5], dates[::5], rotation=45)
plt.xlabel('Date')
plt.ylabel('Number of Patients')
plt.title('Linear Regression Predictions')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
